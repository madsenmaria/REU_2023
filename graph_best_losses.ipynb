### This code takes the runs of the best 6 tuned models and graphs their losses versus their validation losses

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def grapher2(file, epoch_length, end):
    loss = [0]*(epoch_length)
    val_loss = [0]*(epoch_length)
    f = open(file, "r")
    n = 0
               
    ind_l = 0
    ind_vl = 0
    for x in f:
        if len(x.split()) != 0 and x.split()[0] == end:
            # print(type(x))
            if 'val_loss' in x:
                full_str = x.split()
                for i, val in enumerate(full_str):
                    if val == 'loss:':
                        loss[ind_l] =float(full_str[i+1])
                        ind_l = ind_l+1
                    elif val == 'val_loss:':
                        val_loss[ind_vl] =float(full_str[i+1])
                        ind_vl = ind_vl+1
                
                    
            n = n+1
    
    return loss, val_loss 
```

```python
plt.plot(grapher2('Try0931.txt', 30, '100/100')[0], label='Training')
plt.plot(grapher2('Try0931.txt', 30, '100/100')[1], label='Validation')
plt.legend()
plt.ylim([0.055, 0.14])
plt.title('Training and Validation Loss for Tuner:0.0931')
```
